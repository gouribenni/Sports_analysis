```python
from sqlalchemy import types
import datetime
def create_players_table():
    directory_of_files = "/Users/nyzy/data/india_male_json"
    list_files = os.listdir(directory_of_files)
    team_players={}
    for ff in list_files:
        if "json" in ff:
        #print(os.path.join(directory_of_files,ff))
            f = open(os.path.join(directory_of_files,ff))
            
            # returns JSON object as 
            # a dictionary
            data = json.load(f)
            #import pandas as pd
            #pd.DataFrame(players_schema_list)
            player_id_increment = 0
            for each in data["info"]["players"]:
                for xx in data["info"]["players"][each]:
                    if xx not in team_players.keys():
                        team_players[xx]=each
    df = pd.DataFrame(team_players.items(),columns=['player_name','team_name'])
    date_string = "01-01-1980"
    df["date_of_birth"]=datetime.datetime.strptime(date_string,"%m-%d-%Y").date()
    df["batting_style"]="B"
    df["bowling_style"]="B"
    df["player_role"]="B"
    df["player_type"]="B"

    return df



players_table_sql_types =  {'player_name' : types.VARCHAR,
              'team_name':types.VARCHAR,
              'date_of_birth':types.Date,
              'batting_style':types.VARCHAR,
              'bowling_style':types.VARCHAR,
              'player_role':types.VARCHAR,
              'player_type':types.VARCHAR}
df_s = create_players_table()
```
