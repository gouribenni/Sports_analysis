```python
import pandas as pd 
from data_access_layer.flexible_data_read import FlexDataRead
import os 
import constants
list_files = os.listdir(constants.JSON_FILE_DIR)

list_series_dict=[]
for ff in list_files:
        if "json" in ff:
            match_number  = ff.split(".")[0]
            data = FlexDataRead().read_json(os.path.join(constants.JSON_FILE_DIR,ff))
            series_dict = {"series_name":"","series_date":"","season":"","match_id":""}
            for each in (data["info"]["dates"]):
                            #print(data["info"]["event"])
                            series_dict["series_date"]=each
                            try:
                                series_dict["series_name"] = data["info"]["event"]["name"]
                            except:
                                #print(data["info"])
                                series_dict["series_name"] =  data["info"]["teams"][0]+" tour of "+ data["info"]["teams"][1]
                            series_dict["season"] = data["info"]["season"]
                            series_dict["match_id"] = match_number
                            list_series_dict.append(series_dict)
df = pd.DataFrame(list_series_dict)
#df.drop_duplicates(inplace=True)

df_dates = df.groupby(["series_name","season"])["series_date"].aggregate(['min','max']).reset_index()
df_matches = df.groupby(['series_name','season'])['match_id'].agg(match_id_set_text= lambda x: set(x))
df_final = df_matches.merge(df_dates,how='inner',left_on=['series_name','season'],right_on=['series_name',"season"])
#df.columns =["series_name","season","series_start_date","series_end_date"]
#pd.DataFrame.from_dict(series_dict,orient='index').transpose()
df_final["match_id_set_text"]=df_final["match_id_set_text"].apply(lambda x:list(x))
df_final["match_id_set_text"]=df_final["match_id_set_text"].apply(json.dumps)
df_final["winner_id"]=None
#df = df.drop_duplicates()
print(df_final)

```
